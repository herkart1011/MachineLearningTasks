Prompt : dari dataset yang ditampilkan berikan penjelasan terhadap fitur dataset

1. **Gender** : Jenis kelasmin dari siswa, berupa female atau male.

2. **Race/etnicity** : Menunjukkan kelompok ras atau etnis siswa, dengan kategori 'Grup A', 'Grup B', dan 'Grup C'. Dapat digunakan untuk menganalisis variasi dalam kinerja akademis berdasar latar belakang etnis.

3. **Parental level of education** : Menunjukkan tingkat pendidikan orang tua siswa, seperti 'high school', 'bachelor's degree', 'master's degree'. Ini penting untuk memahami pengaruh pendidikan orang tua terhadap kinerja siswa

4. **Lunch** : Menunjukkan jenis makan siang yang diterima siswa yang berupa 'standard' atau 'free/reduced'. Dapat mencerminkan status ekonomi siswa dan dampak terhadap pendidikan.

5. **Test Preparation Course** : Menunjukkan apakah siswa telah menyelesaikan kursus persiapan ujian, dengan pilihan 'completed' atau 'none'. Ini berguna untuk menilai bagaimana persiapan akademis dapat mempengaruhi hasil ujian.

6. **Math Score** : Menunjukkan skor yang diperoleh siswa dalam ujian matematika, dalam rentang 0-100. Ini adalah salah satu ukuran kinerja akademis siswa.

7. **Reading Score** : Menunjukkan skor yang diperoleh siswa dalam ujian membaca, juga dalam rentang 0-100. Ini merupakan ukuran kinerja akademis dalam bidang membaca.

8. **Writing Score** : Menunjukkan skor yang diperoleh siswa dalam ujian menulis, dalam rentang 0-100. Ini adalah ukuran kinerja akademis di bidang menulis.

Prompt : dari dataset diatas bandingkan gender dengan math score, reading score, dan writing score. Serta berikan kesimpulannya

```python
# Menghitung nilai matematika berdasarkan gender dan jumlah siswa pada setiap nilai
math_scores_by_gender = df.groupby(['gender', 'math score']).size().reset_index(name='count')

# Menampilkan hasil
math_scores_by_gender.sort_values(by='math score', ascending=True)
```

```python
# Menghitung nilai membaca berdasarkan gender dan jumlah siswa pada setiap nilai
reading_scores_by_gender = df.groupby(['gender', 'reading score']).size().reset_index(name='count')

# Menampilkan hasil
reading_scores_by_gender.sort_values(by='reading score', ascending=True)
```

```python
# Menghitung nilai menulis berdasarkan gender dan jumlah siswa pada setiap nilai
reading_scores_by_gender = df.groupby(['gender', 'writing score']).size().reset_index(name='count')

# Menampilkan hasil
reading_scores_by_gender.sort_values(by='writing score', ascending=True)
```

```python
# Menghitung rata-rata skor berdasarkan gender
average_scores_by_gender = df.groupby('gender')[['math score', 'reading score', 'writing score']].mean().reset_index()

# Visualisasi perbandingan skor berdasarkan gender
plt.figure(figsize=(12, 6))

# Bar plot untuk rata-rata skor
average_scores_by_gender.plot(x='gender', kind='bar', legend=True, color=['blue', 'green', 'orange'])
plt.title('Average Scores by Gender')
plt.xlabel('Gender')
plt.ylabel('Average Score')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.legend(title='Subjects', labels=['Math Score', 'Reading Score', 'Writing Score'])
plt.tight_layout()
plt.show()
```

### **Kesimpulan**
* Matematika : Meskipun terdapat perbedaan, antar kedua gender memiliki rata - rata skor yang mendekati satu sama lain, dengan perempuan sedikit lebih unggul
* Reading : Skor rata - rata membaca memiliki perbedaan yang terlihat jelas, dimana skor perempuan lebih tinggi daripada laki- laki.
* Writing : PAda mata pelajaran menulis juga terlihat bahwa perempuan memiliki rata - rata skor yang lebih tinggi dibandingkan dengan laki - laki

```python
# Set the style for seaborn
sns.set(style="whitegrid")

# Create a figure for two subplots side by side and one below
fig, axes = plt.subplots(2, 2, figsize=(16, 10))
fig.suptitle('Scores by Parental Education Level', fontsize=16)

# 1. Math Scores by Parental Education Level
sns.boxplot(x='parental level of education', y='math score', data=df, ax=axes[0, 0], palette='Set2')
axes[0, 0].set_title('Math Scores by Parental Education Level')
axes[0, 0].set_ylabel('Math Score')
axes[0, 0].set_xlabel('Parental Education Level')
axes[0, 0].tick_params(axis='x', rotation=45)

# 2. Reading Scores by Parental Education Level
sns.boxplot(x='parental level of education', y='reading score', data=df, ax=axes[0, 1], palette='Set2')
axes[0, 1].set_title('Reading Scores by Parental Education Level')
axes[0, 1].set_ylabel('Reading Score')
axes[0, 1].set_xlabel('Parental Education Level')
axes[0, 1].tick_params(axis='x', rotation=45)

# 3. Writing Scores by Parental Education Level (This will be the single plot below)
sns.boxplot(x='parental level of education', y='writing score', data=df, ax=axes[1, 0], palette='Set2')
axes[1, 0].set_title('Writing Scores by Parental Education Level')
axes[1, 0].set_ylabel('Writing Score')
axes[1, 0].set_xlabel('Parental Education Level')
axes[1, 0].tick_params(axis='x', rotation=45)

# Hide the empty subplot
axes[1, 1].axis('off')

plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

**Kesimpulan**
* Pada boxplot diatas dengan perbandingan nilai matematika, nilai membaca, dan nilai menulis menunjukkan bahwa siswa dengan pendidikan orang tua yang tinggi seperti "Master Degree" dan "Bachelor Degree" memiliki nilai lebih baik dibandingkan dengan pendidikan orang tua "some high school".Dimana ini menunjukkan pendidikan orang tua dapat mempengaruhi kinerja akademis dari para siswa.
* Pada nilai matematika kelompok dengan pendidikan "some high school" memiliki kinerja skor yang lebih rendah dan memiliki beberapa outlier yang menandakan bahwa ada siswa yang kinerjanya jauh di bawah rata - rata
* Pada boxplot 'writing score' kelompok "some college" dan "some high school" menunjukkan distribusi yang lebih lebar menandakan perbedaan kemampuan menulis antar siswa dari latar belakang pendidikan orang tua yang berbeda.
